6. For the vectors: $\vec{a} = 3\hat{i}-2\hat{j}+\hat{k}$, $\vec{b} = -7\hat{i}+2\hat{j}+3\hat{k}$ . $\newline$ Compute the $\newline$ i) cross product $\newline$ ii) find the angle between the vectors, $\newline$ iii) dot product.

```python
a = np.array([3,-2,1])
b = np.array([-7,2,3])

print("Cross-Product: ",np.cross(a,b),"\n",
"Dot Product: ",np.dot(a,b),'\n',
"Angle Between the vectors(in radian):",np.arccos(np.dot(a,b)/(np.linalg.norm(a)*np.linalg.norm(b))))
```

```python
A = np.zeros((5,5))
B = np.ones((5,5))
C = np.resize(np.random.random(25),(5,5))
D = np.resize(np.random.random(25),(5,5))
```

```python
A = A+10
print("i. \n A: ",A)
B  = B*0.5
print("B:",B)
E = A@C
print("ii.","\n",E,"\n")
```

9. Reading images. Download an image from internet. Read the image into a variable, display
the image type. Convert the image to grayscale (if not so). Display the image, find the
image size.
Perform the following operations on the original image:$\newline$
i)rotate by 30 degree, $\newline$ ii) translate
by [10, 20]. $\newline$ iii) scale the image to 1⁄2 its size and display the image size. Write the matrix
forms and perform the operations (lectures 3-4).

```python
import numpy as np
import skimage.io
from skimage.color import rgb2gray
import matplotlib.pyplot as plt


# Read the image
img = skimage.io.imread('https://www.sharpsightlabs.com/datasets/images/Lamborghini_Aventador_2018.jpg')
```

```python
grey_img =  rgb2gray(img)
plt.imshow(grey_img, cmap = 'gray')
plt.show()
```

```python
from skimage.transform import resize

new_pic = resize(img, (914,1368))

from skimage.io import imshow, show

imshow(new_pic)

show()
```

```python
from skimage.transform import EuclideanTransform
from skimage.transform import warp

matrix =np.array([[1,0,10],[0,1,20],[0,0,1]])

t_form = EuclideanTransform(matrix)

tf_img = warp(img, t_form.inverse)

imshow(tf_img)

show()
```

```python
from skimage.transform import rotate

new_pic = rotate(img, 30)

from skimage.io import imshow, show

imshow(new_pic)

show()
```
